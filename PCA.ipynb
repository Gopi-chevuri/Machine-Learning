**Dimensionality Reduction Using PCA**

import libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

load iris dataset

```python
from sklearn.datasets import load_iris
```

```python
dfs = load_iris()
```

```python
dfs.keys()
```

```python
print(dfs['DESCR'])
```

iris data set

```python
df = pd.DataFrame(dfs['data'],columns=dfs['feature_names'])
```

```python
df.head()
```

normalize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(df)
```

```python
scaler = sc.transform(df)
print(scaler)
```

pca on iris data set

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
pca.fit(scaler)
```

```python
pca_1 = pca.transform(scaler)
```

```python
scaler.shape
```

```python
pca_1.shape
```

```python
pca_1
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_1[:,0],pca_1[:,1],c=dfs['target'],cmap='plasma')
```
